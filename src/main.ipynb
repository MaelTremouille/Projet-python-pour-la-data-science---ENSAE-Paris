On doit d'abord s'assurer d'être au bon endroit et d'avoir le bon *working directory*, à savoir : *Projet-python-pour-la-data-science---ENSAE-Paris*.

```python
import os

def set_working_directory(target_dir_name):
    current_dir = os.getcwd()
    
    while os.path.basename(current_dir) != target_dir_name:
        parent_dir = os.path.dirname(current_dir)
        # Si on atteint la racine sans trouver le dossier cible, on arrête pour éviter une boucle infinie
        if current_dir == parent_dir:  
            raise FileNotFoundError(f"Le dossier cible '{target_dir_name}' n'a pas été trouvé dans l'arborescence.")
        current_dir = parent_dir  # Remonter d'un niveau

    os.chdir(current_dir)
    print(f"Le working directory est maintenant :\n {os.getcwd()}")

# Exemple d'utilisation
try:
    set_working_directory("Projet-python-pour-la-data-science---ENSAE-Paris")
except FileNotFoundError as e:
    print(e)
```

Le répertoire de travail a été correctement défini. Cela garantit que tous les chemins relatifs pour l’importation de modules, le chargement de données et l’enregistrement des résultats pointeront vers les bons fichiers du projet. Cela permet d’éviter les erreurs de chargement de données.

Ces coefficients indiquent l'effet des variables explicatives sur la probabilité d'appartenance à chaque classe, par rapport à une classe de référence.

**Le signe du coefficient (positif ou négatif)**

Un coefficient positif pour une variable donnée (ex. Taux de sel) et une classe donnée (ex. a) signifie qu’une hausse de cette variable augmente la probabilité (ou plus précisément le log-odds) de prédire la classe en question (ici Nutriscore a).

Un coefficient négatif indique au contraire qu’un accroissement de cette variable rend la classe correspondante moins probable.

**L’ampleur (la valeur absolue) du coefficient**

Plus la valeur (en positif ou en négatif) est grande, plus la variable exerce une influence marquée (en faveur ou en défaveur) sur la probabilité de cette classe.

Par exemple, si pour la classe e, le coefficient associé au « Taux de matières grasses saturées » est très positif, cela veut dire qu’un produit riche en graisses saturées est plus susceptible d’être classé en « e » (toutes choses égales par ailleurs).

***Remarque***: Les coefficients se lisent dans l’échelle des « log-odds » (logarithme du rapport de chances). L’effet réel sur la probabilité n’est pas linéaire mais exponentiel. Cependant, l’interprétation « positif = augmente la proba, négatif = diminue la proba » reste la même.
